```python
import numpy as np
import pandas as pd
import scipy 
import matplotlib.pyplot as plt
```

```python
arr = np.loadtxt('ASTR19_S22_group_project_data.txt', skiprows=3, dtype='str')
def np_to_list(arr):
  nums = []
  for elem in arr:
    t1 = int(elem[0])
    t2 = elem[1]
    t3 = float(elem[2])
    nums.append((t1, t2, t3))
  return nums

data = np_to_list(arr)
x_col = []
y_col = []
# print(data)
for d in data:
  tide = d[2]
  day = d[0]
  time = d[1]
  #print(time)
  temp = time.split(':')
  new_time = int(temp[0])*60
  new_time += int(temp[1])
  new_time /= (24*60)
  x_col.append(day+new_time)
  y_col.append(tide)

# print(x_col)
# print(y_col)

X = np.array(x_col)
y = np.array(y_col)

plt.plot(X,y, marker='o')
plt.xlabel('Days Since Start of Year')
plt.ylabel('Tide Height Change')
plt.show()
```
