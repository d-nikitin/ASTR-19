```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1, 100)

def sin(x):
  return np.sin(x)
def cos(x):
  return np.cos(x)
```

```python
y1 = sin(x)
y2 = cos(x)

fig, (p1, p2) = plt.subplots(1, 2)
p1.plot(x, y1)
p1.set_title('sin(x)')
p2.plot(x, y2)
p2.set_title('cos(x)')

plt.show()

fig.savefig('plot.pdf')
```
